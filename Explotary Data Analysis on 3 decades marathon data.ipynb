```python
# importing all the important libraries
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# since our data file is too large therefore we extract minimum data fro our file

# Read the CSV file into a DataFrame
df = pd.read_csv(r"E:\marij perosnal projects\Python Projects new\TWO_CENTURIES_OF_UM_RACES.csv")

# Define the conditions for filtering
condition = ((df['Event distance/length'].str.endswith('50km') | df['Event distance/length'].str.endswith('50mi')) & (df['Year of event'] == 2020))

# Apply the conditions directly for filtering
df2 = df[condition]

df2
```

```python
# let's see how our data looks like

df2.head(5)
```

```python
# we want just abbreviations in our event name column like USA , CHI etc
df2[["Event_name","event_host"]]=df2["Event name"].str.split("(",1,expand=True)
```

```python
# cleaninh up the event_host column
df2["event_host"]=df2["event_host"].str.replace(")","")
```

```python
# taking those rows from df2 where event_host is equal to USA

df2=df2[df2["event_host"]=='USA']
```

```python
# dropping the null values
df2=df2.dropna()
```

```python
#dropping the unrequired columns
df2=df2.drop(columns=["Event name","Athlete club","Athlete country","Athlete age category","event_host"],axis=1)
```

```python
# fixing the data types of specific columns

df2["Athlete year of birth"]=df2["Athlete year of birth"].astype(int)
df2["Athlete average speed"]=df2["Athlete average speed"].astype(float)
```

```python
# finding the age of athletes in 2020

df2["athlete_age"]=2020-df2["Athlete year of birth"]
```

```python
df2.head()
```

```python
# cleaning up the athlete perfromance column
df2["Athlete performance"]=df2["Athlete performance"].str.split(' ').str.get(0)
```

```python
# removing more columns that we are not going to use
df2=df2.drop(columns="Athlete year of birth")
```

```python
#df2=df2.reset_index(drop=True)
```

```python
# renaming the columns

df2=df2.rename(columns={'Year of event':'Year','Event dates':'race_day','Event distance/length':'race_distance',
                       'Event number of finishers':'race_finishers','Athlete performance':'Athlete_performance',
                       'Athlete gender':'Athlete_gender','Athlete average speed':'Athlete_average_speed',
                       'Athlete ID':'Athlete_ID'})
```

```python
# Questions that i have to ask from the data
# difference in average speed on race distance and gender

df2.groupby(["race_distance","Athlete_gender"])["Athlete_average_speed"].mean().round(2)
```

```python
# find top 10 age groups who performed best(have high average speed) in 50mi race. note select 10 age_groups who are older than 20

age_group=df2[(df2["race_distance"]=='50mi') & (df2["athlete_age"] >20)]
best_age_group=age_group.groupby('athlete_age')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
best_age_group.head(10)
```

```python
# find top 10 age groups who performed worst(have low average speed) in 50mi race. note select 10 age_groups who are older than 20

age_group=df2[(df2["race_distance"]=='50mi') & (df2["athlete_age"] >20)]
worst_age_group=age_group.groupby('athlete_age')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True)
worst_age_group.head(10)
```

```python
# finding the effect of seasons on the average speed of athletes

#since there is no seasons column we are going to create one
# fist creating the month column

df2["race_month"]=df2["race_day"].str.split(".").str.get(1).astype(int)

# so the seasons criteria is given below
#spring 3-5
#summer 6-8
#winter 12-2
#autumn 9-11
```

```python
# creating season name column 
df2["season_name"]=df2["race_month"].apply(lambda x:'Winter' if x >11 else 'autumn' if x >8 else "summer" if x >5 else "spring" if x>2 else "Winter")
```

```python
# now finding our result
df2.groupby('season_name')["Athlete_average_speed"].agg(["mean","count"]).sort_values('mean',ascending=False)
```
